```python
"""merge() connects columns or indexes in DataFrame based on one or more keys.
parameters: [left,right,how=,on=,left_on=,right_on=,left_index=,right_index=,sort=,suffixes=(''),copy=,indicator=,validate=]
left: which DataFrame on left side
right: whc=ich DataFrame on right side
on: key based on which merge takes place, input as label, or list
[how={'left','right','outer','inner'} default: inner
left: use only keys from left frame, preserve key order
right: use only keys from right frame, preserve key order
outer: use union
inner: use intersection]

"""
df23 = pd.DataFrame({'ID':[1,2,3,5],'Class':[9,10,11,12]})
df24 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D']})
print(pd.merge(df24,df23,on='ID'),'\n')
print(pd.merge(df24,df23,on='ID',how='outer'),'\n')
print(pd.merge(df24,df23,on='ID',how='left'),'\n')
print(pd.merge(df24,df23,on='ID',how='right'),'\n')
```

```python
pd.merge(df24,df23,indicator=True,how='outer')
```

```python
df25 = pd.DataFrame({'ID':[5,6,7,8],'Name':['A','B','C','D']})
print(pd.merge(df23,df25,left_index=True,right_index=True),'\n')
print(pd.merge(df23,df25,left_index=True,right_index=True,suffixes=('_Higher','_Lower')))


#you cannot just pass simply use left_index, you have to use both left_index and right_index
```

```python
"""Pandas concat function provides various facilities for easily combining together Series, DataFrame, and Panel objects.
numpy also has concat to concat ndarray
Parameters of pd.concat() 
objs, axis=,join=,join_axes=,ignore_index=,keys=,levels=,names=,verify_integrity=,sort=,copy=

"""
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([5,6,7,8])
pd.concat([sr1,sr2])
d1 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D']})
d2 = pd.DataFrame({'ID':[4,6,7,8,9],'Name':['E','F','G','H','I']})
print(pd.concat([d2,d1]),'\n') #it is in order it is written
print(pd.concat([d1,d2],axis=1),'\n')
#But see in both cases indexes are repeated, if we have to make new indexes
print(pd.concat([d1,d2],ignore_index=True))
```

```python
#here see even if indexes are not available, still it is concatenating with NaN
#use join, its default here is 'outer'
print(pd.concat([d1,d2],axis=1,join='inner'),'\n')
print(pd.concat([d1,d2]),'\n')
#you may use keys to label the levels,levels correspond to DataFrame here
print(pd.concat([d1,d2],keys=['d1','d2']),'\n')
```

```python
pd.concat([d1,d2],keys=['d1','d2'],axis=1)
```

```python
"""df_name.join()
It is a convenient method for combining the columns with two potentially differently-indexing.
other, on=,how=,lsuffix=,rsuffix=,sort=
other: DataFrame,series
"""
d3 = pd.DataFrame({'A':[1,2,3],'B':[10,20,30]})
d4 = pd.DataFrame({'C':[4,5,6],'D':[40,50,60]})
d5 = pd.DataFrame({'A':[1,2],'B':[10,20]},index=['a','b'])
d6 = pd.DataFrame({'A':[4,5,6],'D':[40,50,60]})
d3.join(d4)
```

```python
d4.join(d5)
```

```python
d4.join(d5,how='right')
#outer,inner,left,right for how
```

```python
#lsuffix: suffix used from left frame's overlapping columns
#rsuffix: suffix used from right frame's overlapping columns
print(d3.join(d6,lsuffix='_1'),'\n')
print(d3.join(d6,rsuffix='_1'),'\n')
```
